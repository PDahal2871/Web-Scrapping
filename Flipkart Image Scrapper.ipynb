```python
"""
['pant','shirt', 'tshirt', 'jacket', 'sweater', 'dhoti', 'sweatshirt', 'shorts',
             'jeans', 'trousers', 'sherwani']
"""
cloth_list = ['trousers', 'sherwani']
```

```python
new_cloth_list = []
for cloth in cloth_list:
    word = cloth.replace(' ', '+')
    new_cloth_list.append(word)
```

```python
cloth_details_urls = set()
```

```python
"""
Merging all the details of items
"""
options = webdriver.ChromeOptions()
options.add_argument("headless") #to stop opening new chrome browser on every hit
driver=webdriver.Chrome('D:/chromedriver.exe', chrome_options=options) #download chromedriver and give the location
img_urls = []

link_url = 'https://www.flipkart.com'

for i, val in enumerate(new_cloth_list):
    cloth_num = 0
    cloth_name = cloth_list[i]
    for i1 in range(1,25): #iterating though 25 pages (1 page contains around 24 items)
        url = "https://www.flipkart.com/search?q="+val+"&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
        driver.get(url+str(i1))
        time.sleep(1)
        bs = BeautifulSoup(driver.page_source, 'html.parser')
        
        rows = bs.find_all("div", {"class": "_13oc-S"})
        
        if len(rows)<=0: #check if the elements of page is empty
            break

        for row in rows:
            print(cloth_name)
            imgs = row.find_all("div",{"class":"_3ywSr_"})
            
            for image in imgs:
                cloth_num += 1
                img_url = image.find('img')['src']
                
                if img_url in cloth_details_urls: #To remove the duplicate
                    continue
                    
                cloth_details_urls.add(img_url)
                req = requests.get(img_url, allow_redirects=True)
                open(f'Flipkart Images/{cloth_name}/{cloth_num}.jpg', 'wb').write(req.content)
```
