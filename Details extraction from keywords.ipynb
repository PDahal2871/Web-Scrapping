```python
import time
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument("headless") #to stop opening new chrome browser on every hit
driver=webdriver.Chrome('D:/chromedriver.exe', chrome_options=options) #download chromedriver and give the location

url = "https://www.flipkart.com/search?q=" + 'red+checked+shirt' + "&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
driver.get(url+str(1))
time.sleep(3)
bs = BeautifulSoup(driver.page_source, 'html.parser')
img_url = []
names = []
prices = []
product_links = []

link_url = 'https://www.flipkart.com'
rows = bs.find_all("div", {"class": "_13oc-S"})

for row in rows:
    imgs = row.find_all("div",{"class":"_3ywSr_"})
    for image in imgs:
        img_url.append(image.find('img')['src'])
        
    name = row.find_all("a", {"class": "IRpwTa _2-ICcC", "class":"IRpwTa"})
    for n in name:
        names.append(n.getText())
        
    price = row.find_all("div", {"class":"_30jeq3"})
    for price in price:
        prices.append(price.getText())
        
    
    product_link = row.find_all("a", {"class":"IRpwTa"})
    for pl in product_link:
        product_links.append(link_url+pl['href'])
        
print(len(img_url))
print(img_url[1])
print(len(names))
print(names[1])
print(len(prices))
print(prices[1])
print(len(product_links))
print(product_links[1])
```

```python
"""
Merging all the details of items from first 10 videos
"""
options = webdriver.ChromeOptions()
options.add_argument("headless") #to stop opening new chrome browser on every hit
driver=webdriver.Chrome('D:/chromedriver.exe', chrome_options=options) #download chromedriver and give the location
cloth_details = []
img_urls = []
prices = []
names = []
postID = []
cdnURL = []
product_links = []

link_url = 'https://www.flipkart.com'

for i, val in enumerate(new_df['Clothes Details'].values.tolist()):
    for i1 in range(1,6): #iterating though 5 pages (1 page contains around 24 items)
        url = "https://www.flipkart.com/search?q="+val+"&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
        driver.get(url+str(i1))
        time.sleep(1)
        bs = BeautifulSoup(driver.page_source, 'html.parser')
        
        rows = bs.find_all("div", {"class": "_13oc-S"})
        
        if len(rows)<=0: #check if the elements of page is empty
            break

        for row in rows:
            imgs = row.find_all("div",{"class":"_3ywSr_"})
            name = row.find_all("a", {"class": "IRpwTa _2-ICcC", "class":"IRpwTa"})
            price = row.find_all("div", {"class":"_30jeq3"})
            product_link = row.find_all("a", {"class":"IRpwTa"})
            
            total_len = [len(imgs), len(name), len(price), len(product_link)]
            
            if len(set(total_len)) != 1:
                continue
            
            for image in imgs:
                img_urls.append(image.find('img')['src'])
                cloth_details.append(val)
                postID.append(new_df.loc[i, "PostID"])
                cdnURL.append(new_df.loc[i, "cdnURL"])

            
            for n in name:
                names.append(n.getText())

            
            for p in price:
                prices.append(p.getText())
                
            
            for pl in product_link:
                product_links.append(link_url+pl['href'])
            
```
