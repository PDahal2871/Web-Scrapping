```python
from PyDictionary import PyDictionary
from textblob import TextBlob
from textblob import Word
 
keywords_list = []
a ='ronaldo'
text_word = Word(a)
 
synonyms = set()
for synset in text_word.synsets:
    for lemma in synset.lemmas():
        synonyms.add(lemma.name())
         
if len(list(synonyms)) == 0:
    keywords_list.append(a)
else:
    keywords_list.append(a)
    keywords_list.extend(list(synonyms))

dictionary=PyDictionary()

answer = dictionary.synonym(a)
if answer is None:
    keywords_list.append(a)
else:
    keywords_list.append(a)
    keywords_list.extend(answer)
    
print(list(set(keywords_list)))
```
